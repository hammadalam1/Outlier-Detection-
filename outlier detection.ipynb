```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.stats import zscore

data = np.array([3, 6, 2, 9, 5, 6, 6, 13, 11, 1, 16,
14, 41,56])
```

```python
# Calculate quartiles and IQR

Q1 = np.percentile(data, 25)

Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

# Identify outliers using IQR method

iqr_threshold = 1.5

outliers_iqr = np.where((data < (Q1 - iqr_threshold * IQR)) | (data > (Q3 + iqr_threshold *
IQR)))[0]

# Visualize outliers

plt.figure(figsize=(8, 6))

plt.scatter(range(len(data)), data, c='blue', label='Inliers')

plt.scatter(outliers_iqr, data[outliers_iqr], c='red', label='Outliers')

plt.title("Outlier Detection using IQR")

plt.legend()

plt.show()
```

# zscore


```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.stats import zscore

# Given data

data = np.array([3, 6, 2, 9, 5, 6, 6, 13, 11, 1, 16,
14, 41,56])
```

```python
# Calculate Z-scores

z_scores = zscore(data)

# Identify outliers using a Z-score threshold (e.g., 2 or 3)

z_threshold = 2

outliers_zscore = np.where(np.abs(z_scores) > z_threshold)[0]

# Visualize outliers

plt.figure(figsize=(8, 6))

plt.scatter(range(len(data)), data, c='blue', label='Inliers')

plt.scatter(outliers_zscore, data[outliers_zscore], c='red', label='Outliers')

plt.title("Outlier Detection using Z-Score")

plt.legend()

plt.show()
```
